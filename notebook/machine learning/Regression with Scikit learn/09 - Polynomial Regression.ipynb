## Polynomial Regression
* Sometimes relationship between variables & target is of higher polynomial degree
* Transformer can be used to convert data to higher degree
* Linear models can predict coef of these higher degree polynomials

```python
from sklearn.datasets import make_circles

X,y = make_circles(n_samples=1000, noise=.04)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=y,s=10)
```

### linear regression on the above data

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X,y)
lr.score(X,y)
```

### The score on the training data itself is very less this is called under fitting need to increase the dimensions of the data

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol = PolynomialFeatures(degree=2)
```

```python
X_tf = pol.fit_transform(X)
```

```python
lr.fit(X_tf,y)
lr.score(X_tf,y)
```
